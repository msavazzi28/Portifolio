```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
# Configurações do ChromeDriver (ou FirefoxDriver)
chrome_options = Options()
  # Executa o Chrome em modo headless (sem abrir a janela)
  # Substitua pelo caminho correto para o ChromeDriver

# Iniciar o WebDriver (Chrome neste caso)
driver = webdriver.Chrome(options=chrome_options)

# Abrir a página desejada
driver.get("https://olympics.com/en/sports/")

# Capturar o HTML da página após carregar
html = driver.page_source

# Fechar o navegador
driver.quit()

# Exibir ou trabalhar com o HTML




soup = BeautifulSoup(html, 'html.parser')
```

```python
#Extraindo links dos espoertes e salvando em um csv


# Extração dos links que contêm "/en/paris-2024/sports/"
links = ["https://olympics.com" + a['href'] for a in soup.find_all('a', href=True) if a['href'].startswith('/en/')]

# Salvando os links em um arquivo de texto
with open("links.txt", "w") as file:
    for link in links:
        file.write(link + "\n")

print("Links salvos em 'links.txt'.")

#Retirar os links que vem a mais no arquivo txt
```

```python
#Extraindo fotos dos espoertes e salvando em um csv
sports = []
image_links = []

items = soup.find_all('li', class_='b2p-sports-wrapper__item')

for item in items:
    sport_name = item.find('span', class_='b2p-sports-wrapper__item-title').text.strip()
    image_link = item.find('img')['src']

    sports.append(sport_name)
    image_links.append("https://olympics.com" + image_link)

# Criação do DataFrame
df = pd.DataFrame({
    'Sport': sports,
    'Image Link': image_links
})

# Salvando o DataFrame em um arquivo CSV
df.to_csv('sports_images.csv', index=False)

print("Arquivo 'sports_images.csv' criado com sucesso.")
```

```python
#Extraindo "Results" dos links
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import pandas as pd

# Configurações do ChromeDriver
chrome_options = Options()
driver = webdriver.Chrome(options=chrome_options)

# Abrir o arquivo links.txt e ler todos os links
with open('links.txt', 'r') as file:
    links = file.readlines()

# Lista para armazenar os dados
data = []

# Iterar sobre cada link
for link in links:
    link = link.strip()  # Remover espaços em branco e quebras de linha
    if link:
        print(f"Processando link: {link}")
        # Abrir a página desejada
        driver.get(link)

        # Esperar um tempo para garantir que a página foi carregada completamente
        time.sleep(2)  # Ajuste o tempo conforme necessário

        # Capturar o HTML da página após carregar
        html = driver.page_source

        # Analisar o HTML com BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')
        divs = soup.find_all('div', {'data-cy': 'event-item'})

        # Extrair o nome a partir do link
        nome = link.split('/')[-1]

        # Para cada div encontrada, encontrar a tag <a> dentro dela e extrair o texto
        for div in divs:
            a_tag = div.find('a')
            if a_tag:
                # Adicionar uma linha por resultado
                data.append({
                    'link': link,
                    'nome': nome,
                    'resultados': a_tag.text.strip()
                })

# Fechar o navegador
driver.quit()

# Criar um DataFrame a partir dos dados coletados
df = pd.DataFrame(data)

# Exibir o DataFrame
print(df)

# Opcional: salvar o DataFrame em um arquivo CSV
df.to_csv('resultados.csv', index=False)
```
